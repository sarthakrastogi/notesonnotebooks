##creating objects

```
datetime.datetime.now()
```

```
d = datetime.datetime(2000, 12, 5)
```

##Formatting dates

```
print(d.strftime("%Y/%m/%d"))
print(d.strftime("%d %b %y"))
print(d.strftime("%B %d, %Y"))
print(d.strftime("%Y-%m-%d %H:%M:%S"))
print(d.isoformat())
```

It's also possible to access various attributes of the datetime object directly:

```
print(d.year)
print(d.month)
print(d.day)
```

strptime creates a datetime object from a string.

```
datetime.datetime.strptime("December 25, 2010", "%B %d, %Y")
```

These were all offest naive datetime objects. That means the object is naive to any sort of timezone. A naive datetime object contains no timezone information.

```
naive = datetime.datetime.now()
print(naive.tzinfo)
```

```
import pytz #TO CREATE AN OFFSET AWARE OBJECT

d = datetime.datetime.now()
timezone = pytz.timezone("America/Los_Angeles")
d_aware = timezone.localize(d)
d_aware.tzinfo
```

The timedelta class is useful for finding the difference between two dates or times.

```
import datetime
import pytz

d1 = datetime.datetime(2000, 12, 5, 5, 50)
d2 = datetime.datetime(1992, 6, 25, 18, 30)

indy = pytz.timezone("America/Indianapolis")
d1 = indy.localize(d1)
d2 = indy.localize(d2)

diff = d2 - d1 #diff is a timedelta object
print(diff)
```
