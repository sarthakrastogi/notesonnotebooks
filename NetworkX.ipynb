```python
import networkx as nx
```

# Creating graphs

```python
G = nx.Graph() #undirected by default
V = {'Sarthak', 'Prerna', 'Ayesha', 'Sameer', 'Ayushi', 'Tara'}
E = [('Sarthak', 'Prerna'), ('Sarthak', 'Tara'), ('Tara', 'Ayushi'), ('Sameer', 'Ayesha')]

G.add_nodes_from(V)
G.add_edges_from(E)

G.remove_nodes_from({"Ayesha", "Sameer"})
```

```python
print(G.nodes)
print(G.edges)
print(G.number_of_edges)
print(G.number_of_nodes)
print([G.degree(v) for v in G.nodes])
print([G.neighbors(v) for v in G.nodes])
print(nx.ego_graph(G, "Sarthak"))
```
