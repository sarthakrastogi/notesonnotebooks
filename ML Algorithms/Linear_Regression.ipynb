```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('diabetes.csv')
```

```python
df
```

```python
df = df.fillna(df.mean())
```

```python
df.columns
```

```python
X = df['GDP'].values.reshape(-1, 1)
y = df['percentage expenditure'].values.reshape(-1, 1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
predict_train = model.predict(X_train)
print('\nLife expectancy predicted on training data',predict_train) 
print(predict_train.shape)
```

```python
predict_test = model.predict(X_test)
print('\nExpenditure predicted on test data',predict_test)
```

```python
# coefficients of the trained model
print('\nCoefficient of model :', model.coef_)

# intercept of the model
print('\nIntercept of model',model.intercept_)
```

```python
# Root Mean Squared Error on training dataset

from sklearn.metrics import mean_squared_error

rmse_train = mean_squared_error(y_train,predict_train)**(0.5)
print('\nRMSE on train dataset : ', rmse_train)
```

A fitting line can be plotted over the scatter graph:

```python
train_x = np.asanyarray(df[['GDP']])
train_y = np.asanyarray(df[['percentage expenditure']])

plt.scatter(df['GDP'], df['percentage expenditure'],  color='blue')
plt.plot(train_x, model.coef_[0][0]*train_x + model.intercept_[0], '-r')
plt.xlabel("Engine size")
plt.ylabel("Emission")
plt.show()
```
